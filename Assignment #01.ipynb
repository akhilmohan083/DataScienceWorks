# visualizations using the dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpgData.dtypes
```

```python
mpgData[mpgData["horsepower"]=="?"].index
```

```python
mpgDataInInt = mpgData.drop([32, 126, 330, 336, 354, 374])
```

```python
mpgDataInInt["horsepower"]=mpgDataInInt["horsepower"].astype(float)
plt.figure()
plt.scatter(mpgDataInInt['horsepower'], mpgDataInInt['mpg']) 
plt.title('Scatter plot Horsepower Vs mpg') 
plt.xlabel('Horse Power of Engine') 
plt.ylabel('Mileage per gallon')
```

from the above graph its visible that mileage reduces with increase in horsepower

```python
plt.figure()
plt.scatter(mpgDataInInt['horsepower'], mpgDataInInt['displacement']) 
plt.title('Scatter plot Horsepower Vs displacement') 
plt.xlabel('Horse Power of Engine') 
plt.ylabel('Displacement')
```

From the above graph its visible that displacement increase with increase in Horsepower

```python
plt.figure()
plt.scatter(mpgDataInInt['weight'], mpgDataInInt['mpg']) 
plt.title('Scatter plot weight Vs MPG') 
plt.xlabel('weight of vehicle') 
plt.ylabel('MPG of vehicle')
```

From the above graph its clear that mileage reduces with increase in weight of vehicle

```python
plt.figure()
plt.scatter(mpgDataInInt['weight'], mpgDataInInt['displacement'],color="red") 
plt.title('Scatter plot weight Vs Displacement') 
plt.xlabel('weight of vehicle') 
plt.ylabel('Displacement of vehicle')
```

From the above graph its clear that displacement increases with increase in weight of vehicle
