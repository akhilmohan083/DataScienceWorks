# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
carData.dtypes
```

```python
car = carData.groupby(['Car Gender'])['Car Gender'].count()
car.plot(kind='bar', title='Car sales according to gender', ylabel='sales count',
         xlabel='gender', figsize=(6, 5))
```
