# The company wishes to clarify whether there is any increase in sales after stepping into digital marketing

```python
data.describe()
```

```python
# H0 : There is no increase in sales after digital marketing
# Ha : There is increase in sales after digital marketing
from scipy import stats
stats.ttest_rel(data['Sales_After_digital_add(in $)'], data['Sales_before_digital_add(in $)'],alternative='greater')
```

The t statistic value here is 12.09 and p value is less than 0.05
So we are rejecting the H0 , null hypothesis and the result is :  
     There is increase in sales after digital marketing

# The company needs to check whether there is any dependency between the features “Region” and “Manager”

H0: The features region and manager are independent ,  
H1: The features region and manager are dependent

```python
from scipy.stats import chi2_contingency
alpha=.05

data = pd.crosstab(data['Manager'], data['Region'])
Chiscore,p,dof,expected= chi2_contingency(data)
print('Chi score is :',Chiscore)
print('p is :',p)

if p > alpha:
    print('Failed to rejecet Null Hypothesis')
else:
    print('Failed to accept null hypothesis')
```

So we can conclude that there is no dependency between Region and Manager
